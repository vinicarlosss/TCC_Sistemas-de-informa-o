```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
caminho_arquivo = 'StudentPerformanceFactors.csv'
data = pd.read_csv(caminho_arquivo)
num_columns = data.shape[1]
num_rows = data.shape[0]
print(f"Número de colunas: {num_columns}")
print(f"Número de linhas: {num_rows}")
data.head()
```

```python
print(data.info())
print('--> VALORES NULOS POR COLUNA <--')
print(data.isnull().sum())
```

```python
data.describe(include='all')
```

```python
data.hist()
```

```python
# Contar valores únicos por coluna
for col in data.columns:
    print(f"{col}: {data[col].nunique()} valores únicos")
```

```python
# Frequência de categorias para variáveis categóricas
for col in data.select_dtypes(include='object').columns:
    print(f"\n{col}:\n{data[col].value_counts(normalize=True) * 100}")
```

```python
# Correlação entre colunas numéricas
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Distribuição do desempenho escolar

sns.histplot(data['Exam_Score'], bins=10, kde=True)
plt.title('Distribuição do Desempenho Escolar')
plt.xlabel('Desempenho')
plt.ylabel('Frequência')
plt.show()
```

```python
# Exemplo com 'gender' e 'coluna_desempenho'
sns.boxplot(data=data, x='Gender', y='Exam_Score')
plt.title('Desempenho Escolar por Gênero')
plt.show()
```
